```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
dfx = pd.read_csv('xdata.csv')
dfy = pd.read_csv('ydata.csv')
#dfx.head(n=5)
X = dfx.values
Y = dfy.values
Y = Y[ : ,1: ]
X = X[ : ,1: ]
Y = Y.reshape((-1,))
print(Y.shape)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
m = X.shape[0]
print(m)
for i in range(m):
    if Y[i]==1:
        plt.scatter(X[i,0],X[i,1],color='orange')
    else:
        plt.scatter(X[i,0],X[i,1],color='red')

queryX = np.array([2,3])
plt.scatter(queryX[0],queryX[1],marker='*',color='white')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1 - x2)**2))
def knn(X,Y,queryPoint,k=5):
    """1. Find dist of query point from all other points"""
    vals =[]
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    new_vals = np.unique(vals[:,1],return_counts=True)
    
    #print(vals)
    final_prediction = new_vals[1].argmax()
    return final_prediction
```

```python
prediction = knn(X,Y,queryX)
print(prediction)
```
