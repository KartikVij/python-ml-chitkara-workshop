### Implement the Algorithm

```python
def hypothesis(x,theta):
    return theta[0] + x*theta[1]

def error(Y,X,theta):
    m = X.shape[0]
    total_error = 0
    
    for i in range(m):
        e = (Y[i] - hypothesis(X[i],theta))**2
        total_error += e
        
    return 0.5*total_error

def getGrad(X,Y,theta):
    grad0, grad1 = 0,0
    m = X.shape[0]
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad0 += (hx - Y[i])
        grad1 += (hx - Y[i])*X[i]
        
    return (grad0,grad1)
```

### Algorithm - Gradient Descent
    

```python
def gradientDescent(X,Y,learning_rate=0.01):
    theta = np.array([0.0,0.0])
    
    maxItr =200
    itr = 1
    error_list = []
    while(itr<maxItr):
        grad0,grad1 = getGrad(X,Y,theta)
        e = error(Y,X,theta)
        error_list.append(e)
        theta[0] = theta[0] - learning_rate*grad0
        theta[1] = theta[1] - learning_rate*grad1
        print(theta)
        itr += 1
        
    return theta,error_list
```

```python
theta,error_list = gradientDescent(X,Y)
plt.plot(error_list)
plt.show()
```
